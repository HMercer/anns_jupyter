```python
df = pd.read_csv("nuclei.csv")
print(df.info())
df.columns
```

```python
df = df.rename(columns={"D to use": "d", "Z":'z', 'N':'n'})
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

array_length = len(df['z'].to_numpy())
print(np.reshape(df['z'].to_numpy(), (-1, array_length)).shape)

X = df['z'].to_numpy()

Y = df['n'].to_numpy()

Z = df['d'].to_numpy()

x = np.reshape(X, (array_length,-1))
y = np.reshape(Y, (array_length,-1))
z = np.reshape(Z, (array_length,-1))


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
